```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
codes = pd.read_csv("DW3_set_exercise.csv")9
codes
```

```python
#1a
diabetes = []
for ind in codes.index:
    if codes['Diagnosis Code'][ind] in ('E08', 'E09', 'E10', 'E11', 'E13'):
        diabetes.append(codes['Patient ID'][ind])

diabetes_set = set(diabetes)
print(diabetes_set)
```

```python
#1b

print("Cardinality of diabetes set is: ", len(diabetes_set))
```

```python
#2a
covid = []
for ind in codes.index:
    if codes['Diagnosis Code'][ind] in ('U07.1', 'J12.82'):
        covid.append(codes['Patient ID'][ind])
covid_set = set(covid)
print(covid_set)
```

```python
#2b

print("Cardinality of COVID set is: ", len(covid_set))
```

```python
#3a

covid_intersection = []
diabetes_intersection = []

for ind in codes.index:
    if codes['Diagnosis Code'][ind] in ('U07.1', 'J12.82'):
        covid_intersection.append(codes['Patient ID'][ind])
    if codes['Diagnosis Code'][ind] in ('E08', 'E09', 'E10', 'E11', 'E13'):
        diabetes_intersection.append(codes['Patient ID'][ind])
        
covid_intersection_set = set(covid_intersection)
diabetes_intersection_set = set(diabetes_intersection)
        
intersection_set = covid_intersection_set.intersection(diabetes_intersection_set)
print(intersection_set)
```

```python
#3b

print("Cardinality of Intersection set is: ", len(intersection_set))
```

```python
#4a

covid_union = []
diabetes_union = []

for ind in codes.index:
    if codes['Diagnosis Code'][ind] in ('U07.1', 'J12.82'):
        covid_union.append(codes['Patient ID'][ind])
    if codes['Diagnosis Code'][ind] in ('E08', 'E09', 'E10', 'E11', 'E13'):
        diabetes_union.append(codes['Patient ID'][ind])
        
covid_union_set = set(covid_union)
diabetes_union_set = set(diabetes_union)
        
union_set = covid_union_set.union(diabetes_union_set)
print(union_set)
```

```python
print("Cardinality of Union set is: ", len(union_set))
```

```python
#5

from matplotlib_venn import venn2

venn2(subsets = (covid_set, diabetes_set), set_labels = ('COVID', 'Diabetes'))
plt.show()
```

```python
#6a

filtered_codes = codes[codes['Diagnosis Code'].isin(['U07.1', 'J12.82', 'E08', 'E09', 'E10', 'E11', 'E13'])]

covid_dates = {}
diabetes_dates = {}
diabetes_after_covid = set()

for x in filtered_codes.index:
    patient_id = filtered_codes['Patient ID'][x]
    diagnosis_code = filtered_codes['Diagnosis Code'][x]
    date = filtered_codes['Date'][x]
    if diagnosis_code in ('E08', 'E09', 'E10', 'E11', 'E13'):
        diabetes_dates[patient_id] = date
    if diagnosis_code in ('U07.1', 'J12.82'):
        covid_dates[patient_id] = date

for patient_id, date_diabetes in diabetes_dates.items():
    if patient_id in covid_dates and date_diabetes > covid_dates[patient_id]:
        diabetes_after_covid.add(patient_id)  

diabetes_after_covid
```

```python
#6b

print("Cardinality of Diabetes after Covid set is: ", len(diabetes_after_covid))
```

```python
#6c

patient_ids = []
diagnosis_codes = []
dates = []

for x in filtered_codes.index:
    patient_id = filtered_codes['Patient ID'][x]
    diagnosis_code = filtered_codes['Diagnosis Code'][x]
    date = filtered_codes['Date'][x]
    if patient_id in diabetes_after_covid:  
        patient_ids.append(patient_id)
        diagnosis_codes.append(diagnosis_code)
        dates.append(date)

breakdown_df = pd.DataFrame({
    'Patient ID': patient_ids,
    'Diagnosis Code': diagnosis_codes,
    'Date': dates})


diabetes_codes_count = breakdown_df[breakdown_df['Diagnosis Code'].isin(['E08', 'E09', 'E10', 'E11', 'E13'])]
diabetes_code_counts_df = diabetes_codes_count.groupby('Diagnosis Code').size()

print("Count breakdown for each diabetes code occurring only after COVID:")
print(diabetes_code_counts_df)
```
